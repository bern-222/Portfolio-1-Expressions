```python
# Daily Budget Expressions Calculator
# This python program demonstrates the use of variables, arithmetic operations & expressions


print("=========== THE DAILY BUDGET EXPRESSION CALCULATOR ==========")
# Prompts the user for their daily budget   
budget = float(input("Enter your daily budget: ₱"))


# Asks the user for daily expenses
food = float(input("Enter your food expense: ₱"))
transportation = float(input("Enter your transportation expense: ₱"))
supplies = float(input("Enter your school supplies expense: ₱"))


# Addition operator is used to compute the total expenses 
expenses = food + transportation + supplies

 
# Subtraction operator is used to compute the remaining money
remaining = budget - expenses

# Compute the percentage of money spent
# Multiplication and division operators to calculate the money spent
percentage_spent = (expenses / budget) * 100

# Display the results
print("\n========= DAILY BUDGET SUMMARY ==========")
print(f"Total Expenses: ₱{expenses:.2f}")
print(f"Remaining Budget: ₱{remaining:.2f}")
print(f"Percentage Spent: {percentage_spent:.2f}%")
```
